```python
df1 = pd.read_csv("../../1_dataset/1_28descriptors/1_dataset/1_original/dataset.csv",
                  usecols=["InChI",
                          "K3FeCN6", "MesAcr+BF4-", "CuCl",
                          "Diacetyl", "T(p-OMe)PPCo", "Electrode"])
df1.shape
```

```python
df1.columns
```

```python
conditions = [["K3FeCN6", "K$_3$[Fe(CN)$_6$]"],
              ["MesAcr+BF4-", "MesAcr$^+$BF$_4^-$"],
              ["CuCl", "CuCl"],
              ["Diacetyl", "Diacetyl"],
              ["T(p-OMe)PPCo", "T($p$-OMe)PPCo"],
              ["Electrode", "Electrode"]]
```

```python
li_numbers = []
li_on_bits = []
```

```python
for i in range(6):
    condition = conditions[i][0]
    
    print(f"\n{condition}")
    df2 = df1[df1[condition] == 1]

    n = len(df2)
    li_numbers.append(n)

    inchi_list = df2["InChI"].to_list()
    ss = ecfp_metrics.draw_on_bits(inchi_list, f"{condition}.svg")
    li_on_bits.append(ss)
```

```python
li_numbers
```

```python
li_on_bits
```

```python
labels = []

for i in range(6):
    labels.append(f"{conditions[i][1]} ({li_numbers[i]})")
    
labels
```

```python
conditions.append(["AllPhenols"])

n = len(df1)
li_numbers.append(n)

inchi_list = df1["InChI"].to_list()
ss = ecfp_metrics.on_bits(inchi_list, 2, 2048)
li_on_bits.append(ss[0])
```

```python
df_result = pd.DataFrame({"Condition": [e[0] for e in conditions],
                          "Number":li_numbers,
                          "ECFP OnBits": li_on_bits})
df_result
```

```python
df_result.to_csv("ecfp_OnBits.csv", index=False)
```
