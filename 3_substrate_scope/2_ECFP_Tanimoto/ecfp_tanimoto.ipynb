```python
from SSMetrics import ecfp_metrics
```

```python
df1 = pd.read_csv("../../1_dataset/1_28descriptors/1_dataset/1_original/dataset.csv",
                  usecols=["InChI",
                          "K3FeCN6", "MesAcr+BF4-", "CuCl",
                          "Diacetyl", "T(p-OMe)PPCo", "Electrode"])
df1.shape
```

```python
df1.columns
```

```python
conditions = [["K3FeCN6", "K$_3$[Fe(CN)$_6$]"],
              ["MesAcr+BF4-", "MesAcr$^+$BF$_4^-$"],
              ["CuCl", "CuCl"],
              ["Diacetyl", "Diacetyl"],
              ["T(p-OMe)PPCo", "T($p$-OMe)PPCo"],
              ["Electrode", "Electrode"]]
```

```python
li_numbers = []
li_distance_avg = []
li_distance_max = []
li_distance_min = []
li_max_inchi = []
li_min_inchi = []
```

```python
for i in range(6):
    condition = conditions[i][0]
    
    print(f"\n{condition}")
    df2 = df1[df1[condition] == 1]

    n = len(df2)
    li_numbers.append(n)

    inchi_list = df2["InChI"].to_list()
    ss = ecfp_metrics.tanimoto(inchi_list)
    li_distance_avg.append(ss[0])
    li_distance_max.append(ss[1])
    li_distance_min.append(ss[2])
    li_max_inchi.append(ss[3])
    li_min_inchi.append(ss[4])
```

```python
labels = []

for i in range(6):
    labels.append(f"{conditions[i][1]} ({li_numbers[i]})")
    
labels
```

```python
for i in range(6):
    condition = conditions[i][0]
    print(f"\n{condition}")

    for j in range(len(li_max_inchi[i])):
        for k in range(2):
            name = f"max_{j+1}_{k+1}"
            inchi = li_max_inchi[i][j][k]
            print(name, inchi)
            Draw.MolToFile(Chem.MolFromInchi(inchi),
                           f"{condition}_{name}.png",
                           size=(600,300))
            
    for j in range(len(li_min_inchi[i])):
        for k in range(2):
            name = f"min_{j+1}_{k+1}"
            inchi = li_min_inchi[i][j][k]
            print(name, inchi)
            Draw.MolToFile(Chem.MolFromInchi(inchi),
                           f"{condition}_{name}.png",
                           size=(600,300))
```

```python
conditions.append(["AllPhenols"])

n = len(df1)
li_numbers.append(n)

inchi_list = df1["InChI"].to_list()
ss = ecfp_metrics.tanimoto(inchi_list)
li_distance_avg.append(ss[0])
li_distance_max.append(ss[1])
li_distance_min.append(ss[2])
```

```python
df_result = pd.DataFrame({"Condition": [e[0] for e in conditions],
                          "Number":li_numbers,
                          "ECFP Tanimoto_avg": li_distance_avg,
                          "ECFP Tanimoto_max": li_distance_max,
                          "ECFP Tanimoto_min": li_distance_min})
df_result
```

```python
df_result.to_csv("ecfp_tanimoto.csv", index=False)
```
