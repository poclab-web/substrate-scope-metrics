```python
from sklearn.manifold import TSNE
```

```python
df1 = pd.read_csv("../1_dataset/2_standardization/standardized_dataset.csv")
df1.shape
```

```python
df1.columns
```

```python
df2 = df1.loc[:, "m_1_L" : "num_Br"]
df2.shape
```

```python
df2.columns
```

```python
tsne = TSNE(n_components=2, random_state=0)
df2_tsne = tsne.fit_transform(df2)
```

```python
df_score = pd.DataFrame(df2_tsne, columns=["tSNE1", "tSNE2"])
df_score.shape
```

```python
df3 = pd.concat([df1, df_score], axis=1)
df3.shape
```

```python
df3.columns
```

```python
df3.to_csv("tsne.csv", index=False)
```
