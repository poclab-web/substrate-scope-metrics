```python
df1 = pd.read_csv("../../../1_dataset/1_original/dataset.csv")
df1.shape
```

```python
df1["K3FeCN6"].value_counts()
```

```python
df2 = df1[df1["K3FeCN6"] != 0].copy()
df2.shape
```

```python
df2.columns[3]
```

```python
li = []
for i in range(len(df2)):
    if df2.iat[i, 3] == 1:
        li.append("Positive")
    else:
        li.append("Negative")

len(li)
```

```python
df2["K3FeCN6"] = li
df2["K3FeCN6"].value_counts()
```

```python
df2.columns[9]
```

```python
df2.columns[36]
```

```python
for i in range(9, 37):
    fig = plt.figure()
    ax = fig.add_subplot(2, 1, 1)

    sns.swarmplot(x=df2.columns[i],
                  y="K3FeCN6",
                  data=df2,
                  hue="K3FeCN6",
                  palette=[(255/255, 75/255, 0/255),
                           (0/255, 90/255, 255/255)],
                  legend=False)

    ax.set_ylabel("")
    plt.savefig(f"{i-8}_{df2.columns[i]}.png",
                bbox_inches="tight")
    plt.show()
```
