```python
df1 = pd.read_csv("../../../1_dataset/1_original/dataset.csv")
df1.shape
```

```python
df1.columns
```

```python
# K3FeCN6
df_k3fecn6 = df1[df1["K3FeCN6"] == 1].copy()
df_k3fecn6["Condition"] = "K$_3$[Fe(CN)$_6$]"
df_k3fecn6.shape
```

```python
# MesAcr+BF4-
df_mesacr = df1[df1["MesAcr+BF4-"] == 1].copy()
df_mesacr["Condition"] = "MesAcr$^+$BF$_4^-$"
df_mesacr.shape
```

```python
# CuCl
df_cucl = df1[df1["CuCl"] == 1].copy()
df_cucl["Condition"] = "CuCl"
df_cucl.shape
```

```python
# Diacetyl
df_diacetyl = df1[df1["Diacetyl"] == 1].copy()
df_diacetyl["Condition"] = "Diacetyl"
df_diacetyl.shape
```

```python
# T(p-OMe)PPCo
df_cocalt = df1[df1["T(p-OMe)PPCo"] == 1].copy()
df_cocalt["Condition"] = "T($p$-OMe)PPCo"
df_cocalt.shape
```

```python
# Electrode
df_electrode = df1[df1["Electrode"] == 1].copy()
df_electrode["Condition"] = "Electrode"
df_electrode.shape
```

```python
df2 = pd.concat([df_k3fecn6, df_mesacr, df_cucl, df_diacetyl, df_cocalt, df_electrode])
df2.shape
```

```python
df2["Condition"].value_counts()
```

```python
df2.columns[9]
```

```python
df2.columns[36]
```

```python
for i in range(9, 37):
    fig = plt.figure()
    ax = fig.add_subplot(2, 1, 1)

    sns.swarmplot(x=df2.columns[i],
                  y="Condition",
                  data=df2,
                  hue="Condition",
                  hue_order=["K$_3$[Fe(CN)$_6$]",
                             "MesAcr$^+$BF$_4^-$",
                             "CuCl",
                             "Diacetyl",
                             "T($p$-OMe)PPCo",
                             "Electrode"],
                  palette=[(255/255, 75/255, 0/255),
                           (153/255, 0/255, 153/255),
                           (3/255, 175/255, 122/255),
                           (246/255, 170/255, 0/255),
                           (77/255, 196/255, 255/255),
                           (255/255, 128/255, 130/255)],
                           legend=False)

    ax.set_ylabel("")
    plt.savefig(f"{i-8}_{df2.columns[i]}.png",
                bbox_inches="tight")
    plt.show()
```
