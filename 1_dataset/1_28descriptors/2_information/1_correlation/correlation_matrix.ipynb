```python
df1_corr = df1.loc[:, "m_1_L":"num_Br"].corr()
df1_corr.shape
```

```python
df1_corr.to_csv("correlation_matrix.csv")
```

```python
fig = plt.figure(figsize=(25, 10))
ax = fig.add_subplot(111)
sns.heatmap(df1_corr, annot=True, fmt=".3f", center=0, cmap="RdBu_r", ax=ax)
fig.savefig("correlation_matrix.png")
plt.show()
```
