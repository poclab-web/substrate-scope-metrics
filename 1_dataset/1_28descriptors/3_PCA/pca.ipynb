```python
from sklearn.decomposition import PCA
```

```python
df1 = pd.read_csv("../1_dataset/2_standardization/standardized_dataset.csv")
df1.shape
```

```python
df1.columns
```

```python
df2 = df1.loc[:, "m_1_L" : "num_Br"]
df2.shape
```

```python
df2.columns
```

```python
pca = PCA(n_components=2, whiten=True)
pca.fit(df2)
```

```python
df_score = pd.DataFrame(pca.transform(df2), columns=["PC1", "PC2"])
df_score.shape
```

```python
df3 = pd.concat([df1, df_score], axis=1)
df3.shape
```

```python
df3.columns
```

```python
df3.to_csv("pca.csv", index=False)
```

```python
df_corr = df3.loc[:, "m_1_L" : "PC2"].corr()
df_corr.loc[:"num_Br", "PC1":]
```

```python
df_corr["PC1"].sort_values(ascending=False)
```

```python
df_corr["PC2"].sort_values(ascending=False)
```
