```python
from sklearn.preprocessing import StandardScaler
```

```python
df1 = pd.read_csv("../1_original/dataset.csv")
df1.shape
```

```python
df2 = df1.loc[:, "m_1_L" : "num_Br"]
df2.shape
```

```python
df2.columns
```

```python
# Standardization
standardized_df2 = StandardScaler().fit_transform(df2)
standardized_df2
```

```python
df_standardized = pd.DataFrame(standardized_df2, columns=df2.columns)
df_standardized
```

```python
df_standardized.shape
```

```python
df3 = pd.concat([df1.loc[:, "SMILES" : "Electrode"], df_standardized], axis=1)
df3.shape
```
