```python
df1 = pd.read_csv("../1_28descriptors/1_dataset/1_original/dataset.csv",
                  usecols=["SMILES", "InChI", "InChIKey", "K3FeCN6"])
df1.shape
```

```python
df1.columns
```

```python
df1.drop(df1[df1["K3FeCN6"] == 0].index, inplace=True)
df1.shape
```

```python
df1["K3FeCN6"].value_counts()
```

```python
df1["ROMol"] = df1["InChI"].map(Chem.MolFromInchi)
df1.shape
```

```python
calc = Calculator(descriptors, ignore_3D=True)
df_mordred = calc.pandas(df1.loc[:, "ROMol"])
```

```python
df_mordred.shape
```

```python
df_mordred.columns
```

```python
df2 = df_mordred.select_dtypes(include="number")
df2
```

```python
df2.shape
```

```python
df_error = df_mordred.select_dtypes(exclude="number")
df_error
```

```python
df_error.shape
```

```python
df1.drop(columns="ROMol", inplace=True)
df1.columns
```

```python
df3 = pd.merge(df1, df2, left_index=True, right_index=True)
df3
```

```python
df3.shape
```

```python
df3.dropna(axis=1, inplace=True)
df3.shape
```
