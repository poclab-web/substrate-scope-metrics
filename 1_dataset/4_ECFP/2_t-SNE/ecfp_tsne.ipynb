```python
from SSMetrics import ecfp
```

```python
df1 = pd.read_csv("../../1_28descriptors/1_dataset/1_original/dataset.csv",
                  usecols=["InChI",
                          "K3FeCN6", "MesAcr+BF4-", "CuCl",
                          "Diacetyl", "T(p-OMe)PPCo", "Electrode"])
df1.shape
```

```python
df1.columns
```

```python
df2 = ecfp(df1["InChI"])
df2.shape
```

```python
df2.columns
```

```python
tsne = TSNE(n_components=2, random_state=0)
df2_tsne = tsne.fit_transform(df2)
```

```python
df_score = pd.DataFrame(df2_tsne, columns=["tSNE1", "tSNE2"])
df_score.shape
```

```python
df3 = pd.concat([df1, df_score], axis=1)
df3.shape
```

```python
df3.columns
```

```python
df3.to_csv("ecfp_tsne.csv", index=False)
```
