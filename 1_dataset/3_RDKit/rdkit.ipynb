```python
from rdkit import Chem, rdBase
rdBase.rdkitVersion
```

```python
from rdkit.Chem import Descriptors
```

```python
df1 = pd.read_csv("../1_28descriptors/1_dataset/1_original/dataset.csv",
                  usecols=["SMILES", "InChI", "InChIKey", "K3FeCN6"])
df1.shape
```

```python
df1.columns
```

```python
df1.drop(df1[df1["K3FeCN6"] == 0].index, inplace=True)
df1.shape
```

```python
df1["K3FeCN6"].value_counts()
```

```python
df1["ROMol"] = df1["InChI"].map(Chem.MolFromInchi)
df1.shape
```

```python
len(Descriptors.descList)
```

```python
for i, j in Descriptors.descList:
    df1[i] = df1["ROMol"].map(j)

df1.shape
```

```python
df1.columns
```

```python
df1.drop(columns="ROMol", inplace=True)
df1.shape
```

```python
df1.dropna(axis=1, inplace=True)
df1.shape
```

```python
df1.to_csv("rdkit.csv", index=False)
```
