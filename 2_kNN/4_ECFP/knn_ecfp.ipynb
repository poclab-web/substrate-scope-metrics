```python
from SSMetrics import ecfp, loo_knn
```

```python
df1 = pd.read_csv("../../1_dataset/1_28descriptors/1_dataset/1_original/dataset.csv",
                  usecols=["SMILES", "InChI", "InChIKey", "K3FeCN6"])
df1.shape
```

```python
df1.columns
```

```python
df1.drop(df1[df1["K3FeCN6"] == 0].index, inplace=True)
df1.shape
```

```python
df1["K3FeCN6"].value_counts()
```

```python
x = ecfp(df1["InChI"]).to_numpy()
x.shape
```

```python
y = np.array(df1["K3FeCN6"])
y.shape
```

```python
accuracy = []
precision = []
recall = []
f1 = []

for i in range(1, 16):
    classifier = loo_knn(x, y, i)

    accuracy.append(classifier[2])
    precision.append(classifier[3])
    recall.append(classifier[4])
    f1.append(classifier[5])
    
df_knn = pd.DataFrame({"k": list(range(1, 16)),
                       "Accuracy": accuracy,
                       "Precision": precision,
                       "Recall": recall,
                       "F1": f1})
df_knn
```

```python
df_knn.to_csv("knn_ecfp.csv", index=False)
```
