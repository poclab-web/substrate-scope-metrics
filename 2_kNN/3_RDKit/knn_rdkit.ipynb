```python
df2 = df1.iloc[:, 4:]
df2.shape
```

```python
df2.columns
```

```python
x = np.array(df2)
x.shape
```

```python
y = np.array(df1["K3FeCN6"])
y.shape
```

```python
accuracy = []
precision = []
recall = []
f1 = []

for i in range(1, 16):
    classifier = loo_std_knn(x, y, i)

    accuracy.append(classifier[2])
    precision.append(classifier[3])
    recall.append(classifier[4])
    f1.append(classifier[5])
    
df_knn = pd.DataFrame({"k": list(range(1, 16)),
                       "Accuracy": accuracy,
                       "Precision": precision,
                       "Recall": recall,
                       "F1": f1})
df_knn
```

```python
df_knn.to_csv("knn_rdkit.csv", index=False)
```
